```python
from ultralytics import YOLO
import os

# YOLOv8 모델 초기화 (예: yolov8n, yolov8s, yolov8m, yolov8l, yolov8x 등)
model = YOLO('yolov8n.yaml')

# 데이터셋 경로 설정
data_path = 'D:\Github\ML_yolo_v8\dataset\data.yaml'

# 모델 학습
model.train(data=data_path, epochs=50, imgsz=640, batch=16, name='yolov8_toilet_paper')
```

```python
import cv2
from ultralytics import YOLO

# YOLO 모델 로드
model_path = r'D:\Github\ML_yolo_v8\runs\detect\yolov8_toilet_paper3\weights\best.pt'
model = YOLO(model_path)

# 카메라 열기 (0은 기본 카메라, 다른 카메라를 사용하려면 적절한 인덱스로 변경)
cap = cv2.VideoCapture(0)

# 카메라가 정상적으로 열렸는지 확인
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

while True:
    # 카메라 프레임 읽기
    ret, frame = cap.read()

    # 프레임을 읽을 수 없는 경우 루프 탈출
    if not ret:
        break

    # YOLO 모델로 객체 검출
    results = model(frame)[0]  # 결과를 리스트에서 첫 번째 요소로 접근

    # 검출된 객체 그리기
    for result in results.boxes:
        x1, y1, x2, y2 = result.xyxy[0]  # 바운딩 박스 좌표
        conf = result.conf[0]  # 신뢰도
        cls = result.cls[0]  # 클래스 인덱스
        label = f'{model.names[int(cls)]}: {conf:.2f}'
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 결과 프레임 보여주기
    cv2.imshow('YOLO Object Detection', frame)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 카메라 및 윈도우 해제
cap.release()
cv2.destroyAllWindows()
```
